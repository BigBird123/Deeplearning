```
train_size = 1280000
```

```
train_senquences = senquences[:train_size]
train_labels = labels[:train_size]
test_senquences = senquences[train_size:]
test_labels = labels[train_size:]
```

```
import numpy as np 
train_labels = np.array(train_labels)
test_labels = np.array(test_labels)
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
vocab_size = 10000
```

```
tokenizer = Tokenizer(num_words=vocab_size,oov_token="<OOV>") # 0 : thì có bao nhiêu lấy bấy nhiêu
```

```
tokenizer.fit_on_texts(train_senquences)
```

```
tokenizer.word_index #size cua vocab
```

```
data['_TheSpecialOne_'].iloc[:6].count()
```

```
train_senquences =tokenizer.texts_to_sequences(train_senquences)
```

```
print(train_senquences[0:30],type(train_senquences))
```

```
test_senquences = tokenizer.texts_to_sequences(test_senquences)
```

```
len(train_senquences)
```

```
padded_train_sequences = pad_sequences(train_senquences,
                                       maxlen = 60)
```

```
padded_train_sequences.shape
```

```
padded_test_sequences = pad_sequences(test_senquences,
                                       maxlen = 60)
```

```
padded_test_sequences.shape
```

```
from tensorflow.keras.layers import Bidirectional 
from tensorflow.keras.layers import  LSTM
from tensorflow.keras.layers import  GRU
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Embedding
```

```
model = Sequential()
model.add(Embedding(input_dim= vocab_size,output_dim=64)) # 64 là chiều vecter
model.add(Bidirectional(LSTM(64, return_sequences = True)))
model.add(Bidirectional(LSTM(32)))
model.add(Dense(64,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```
model.summary()
```

```
model.compile(loss = 'binary_crossentropy',optimizer='adam',metrics='acc')
model.fit(padded_train_sequences,train_labels, 
          epochs=5,
          batch_size = 512,
          validation_data=(padded_test_sequences,test_labels))
```
