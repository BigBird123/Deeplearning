```
#phải bỏ số .
vocab= []
for i in range(len(df)):
  alist = []
  for item in list(df.iloc[i]):
    if isinstance(item,str):
      alist.append(item)
  vocab.append(alist)
```

```
len(vocab[1])
```

```
vocab
```

```
from gensim.models import  Word2Vec
```

```
model = Word2Vec(vocab, size=16, window=5,workers=3,sg = 0 ) # sg =0 -> CBROW , sg = 1 -> SkipGram
#size của hitden layer chứ ko phải size của vocabury : 16x1 
```

```
model['Toyota Camry']
```

```
#tinh ra goc cua 2 vecter la bao nhieu (cos)
model.similarity("Toyota Camry","Honda")
```

```
model.most_similar('BMW')[:6]
```
